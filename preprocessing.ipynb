## Z-score help to remove outliers from dataset

```python
from scipy.stats import norm 
import numpy as np
plt.hist(df.Length1, bins=15, rwidth=0.8, density=True)
plt.xlabel("price")
plt.ylabel("count")

rng = np.arange(df.Length1.min(),df.Length1.max(),0.05)
plt.plot(rng, norm.pdf(rng, df.Length1.mean(),df.Length1.std()))
```

```python
new_df = df
new_df
```

```python
#zscore for weight column
new_df['zscore'] = (new_df.Weight	 - new_df.Weight.mean())/new_df.Weight.std()
new_df
```

```python
#show outliers
new_df[(new_df['zscore'] > 2) | (new_df['zscore'] < -2)]
```

```python
no_outliers_df = new_df[(new_df['zscore'] < 2) & (new_df['zscore'] > -2)]
no_outliers_df
```

## simple Linear Regression with sklearn

```python
df = pd.read_csv("/content/drive/MyDrive/0. AAA Sec B and Sec C /SEC C/Salary_Data.csv")
df.head(2)
```

```python
x = df.iloc[:,0:1]
x
```

```python
y = df.iloc[:,1:2]
y
```

```python
plt.scatter(x, y, color="green")   
plt.title("Salary vs Experience")  
plt.ylabel("Years of Experience")  
plt.xlabel("Salary")  
plt.show() 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,random_state=42, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg = reg.fit(x_train,y_train)
```

```python
reg.score(x_test, y_test)
```

```python
#now we will Prediction of Test and Training set result  
x_pred= reg.predict(x_train) 
y_pred= reg.predict(x_test)  
 
```

```python
# red is the best fit line 
plt.scatter(x_train, y_train, color="green")   
# plt.plot(x_train, x_pred, color="red")    
plt.title("Salary vs Experience (Training Dataset)")  
plt.xlabel("Years of Experience")  
plt.ylabel("Salary")  
plt.show()  
```

```python
# red is the best fit line 
plt.scatter(x_test, y_test, color="blue")   
plt.plot(x_test, y_pred, color="red")    
plt.title("Salary vs Experience (Testing Dataset)")  
plt.xlabel("Years of Experience")  
plt.ylabel("Salary")  
plt.show() 
```
